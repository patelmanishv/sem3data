### Exercise 1 : Write MRO for A,B,C,D
```PYTHON
class A:
  def method(self):
    print("A.method() called")

class B:
  def method(self):
    print("B.method() called")

class C(A, B):
  pass

class D(C, B):
  pass

```

### Exercise 2 : Write MRO for A,B,C,D
```PYTHON
class A:
   def m1(self):
       print("m1 from A")
class B(A):
   def m1(self):
       print("m1 from B")
class C(A):
   def m1(self):
       print("m1 from C")
class D(B, C):
   def m1(self):
       print("m1 from D")

```

### Exercise 3 : Write MRO for A,B,C,X,Y,P
```PYTHON
class A:
   def m1(self):
       print("m1 from A")
class B:
   def m1(self):
       print("m1 from B")
class C:
   def m1(self):
       print("m1 from C")
class X(A, B):
   def m1(self):
       print("m1 from C")
class Y(B, C):
   def m1(self):
       print("m1 from A")
class P(X, Y, C):
   def m1(self):
       print("m1 from P")
```
