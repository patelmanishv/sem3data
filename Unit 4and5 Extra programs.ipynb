```python
# There are n children standing in a line. Each child is assigned a rating value given in the integer
# array ratings.
# You are giving candies to these children subjected to the following requirements:
# Each child must have at least one candy.
# Children with a higher rating get more candies than their neighbors.
# Return the minimum number of candies you need to have to distribute the candies to the
# children.
# Example 1:
# Input: ratings = [1,0,2]
# Output: 5
def min_candies(ratings):
    n = len(ratings)
    if n == 0:
        return 0
    
    candies = [1] * n  # Each child starts with at least one candy

    # First pass: left to right
    for i in range(1, n):
        if ratings[i] > ratings[i - 1]:
            candies[i] = candies[i - 1] + 1

    # Second pass: right to left
    for i in range(n - 2, -1, -1):
        if ratings[i] > ratings[i + 1]:
            candies[i] = max(candies[i], candies[i + 1] + 1)

    return candies,sum(candies)
```

A transaction is equivalent to (buying + selling) of a stock and new transaction can start only when the previous transaction has been completed.

```python
# Python program to find maximum possible profit with 
# at most two transactions using state machine and dp

def maxProfit(price):
    # Variables to store the maximum profit after the first and second transactions
    first_buy = float('-inf')  
    first_sell = 0      
    second_buy = float('-inf') 
    second_sell = 0      

    # Iterate over each day's stock price
    for p in price:
        # Calculate maximum profit after first buy (negative price to track spending)
        first_buy = max(first_buy, -p)
        
        # Calculate maximum profit after first sell
        first_sell = max(first_sell, first_buy + p)
        
        # Calculate maximum profit after second buy
        second_buy = max(second_buy, first_sell - p)
        
        # Calculate maximum profit after second sell
        second_sell = max(second_sell, second_buy + p)
    
    # The result is the maximum profit after the second sell
    return second_sell

price = [10, 22, 5, 75, 65, 80]
print(price,maxProfit(price))
price = [1,2,3,4,5]
print(price,maxProfit(price))
price = [3,3,5,0,0,3,1,4]
print(price,maxProfit(price))
price = [10, 22, 5, 75, 65, 80]
print(price,maxProfit(price))
price = [90,80,70]
print(price,maxProfit(price)) 
price = [10, 30, 50, 60]
print(price,maxProfit(price)) 
```

>Given an unsorted array of integers, find the length of the longest sequence of consecutive elements. Consecutive elements are those that appear in increasing order without gaps, and the sequence should contain distinct integers. Implement this without using a set.

**Examples:**

- Input: `[100, 4, 200, 1, 3, 2]`  
  Output: `4`  
  Explanation: The longest consecutive sequence is `[1, 2, 3, 4]`.

- Input: `[10, 5, 12, 3, 55, 4, 6, 1, 2]`  
  Output: `6`  
  Explanation: The longest consecutive sequence is `[1, 2, 3, 4, 5, 6]`.

**Constraints:**
- The array can contain both positive and negative integers.
- The length of the array is at least 1.
- You are required to solve this without using a set.

```python
def longestConsecutive(nums):
    if not nums:
        return 0

    # Sort the array
    nums.sort()
    longest_streak = 1
    current_streak = 1

    for i in range(1, len(nums)):
        # Check if the current number is consecutive
        if nums[i] == nums[i - 1] + 1:
            current_streak += 1
        elif nums[i] != nums[i - 1]:  # Skip duplicates
            longest_streak = max(longest_streak, current_streak)
            current_streak = 1

    # Final comparison to catch the longest streak at the end
    return max(longest_streak, current_streak)


# Example usage
arr = [-3,-2,-1,0,1,2,100, 40, 200, 1, 3, 20]
print(longestConsecutive(arr))  # Output: 4, as the longest consecutive sequence is [1, 2, 3, 4]
```

>Given an IP address as input, write a Python program to check whether the given IP Address is Valid or not.
What is an IP (Internet Protocol) Address? 
Every computer connected to the Internet is identified by a unique four-part string, known as its Internet Protocol (IP) address. An IP address (version 4) consists of four numbers (each between 0 and 255) separated by periods. The format of an IP address is a 32-bit numeric address written as four decimal numbers (called octets) separated by periods; each number can be written as 0 to 255 (E.g. – 0.0.0.0 to 255.255.255.255).


```python
# Function to check whether a string is a valid part of an IP address
def is_valid_part(s):
    # The part should be a digit and its length should not exceed 3
    if len(s) > 3 or not s.isdigit():
        return False
    
    # Convert string to an integer and check if it's within the valid range (0 to 255)
    num = int(s)
    return 0 <= num <= 255

# Function to check if the given string represents a valid IP address
def is_valid_ip(ip_str):
    # Return False if the input is None or an empty string
    if not ip_str:
        return False

    # Split the input string into parts based on the '.' delimiter
    parts = ip_str.split('.')

    # Check if there are exactly 4 parts (the IP address should have 3 dots)
    if len(parts) != 4:
        return False

    # Validate each part of the IP address
    for part in parts:
        if not is_valid_part(part):
            return False

    # If all parts are valid, return True
    return True

# Driver code to test the function with different IP addresses
ip_addresses = [
    "128.0.0.1",
    "125.16.100.1",
    "125.512.100.1",  # Invalid part '512'
    "125.512.100.abc",
    "255.255.255.255" # Invalid part 'abc'
]

# Check and print whether each IP is valid or not
for ip in ip_addresses:
    print(f"{ip}: {'Valid' if is_valid_ip(ip) else 'Not valid'}")
```

```python
### shortest palindrome
# You are given a string s. You can convert s to a palindrome by adding characters in front of it.
# Return the shortest palindrome you can find by performing this transformation.
# Example 1:
# Input: s = "aacecaaa"
# Output: "aaacecaaa"
# Input: s = "abcd"
# Output: "dcbabcd"
def shortestPalindrome(s: str) -> str:
    if not s:
        return s
    
    # Find the longest palindromic prefix
    right = len(s) - 1
    while right >= 0:
        if s[:right + 1] == s[:right + 1][::-1]:  # Check if the prefix is a palindrome
            break
        right -= 1
    
    # The non-palindromic suffix to add in reverse order at the beginning
    suffix = s[right + 1:]
    return suffix[::-1] + s
```

```python
def compress_string(s: str) -> str:
    # Initialize the result string
    compressed = ''
    
    # Initialize the pointer to track the index in the string
    i = 0
    n = len(s)
    
    # Iterate through the string
    while i < n:
        # Current character
        char = s[i]
        count = 1
        
        # Count the number of consecutive occurrences of the character
        while i + 1 < n and s[i + 1] == char:
            i += 1
            count += 1
        
        # Append the character and count (only if count > 1)
        compressed += char
        if count > 1:
            compressed += str(count)
        
        # Move to the next character
        i += 1
    
    return compressed

# Sample Test Cases
print(compress_string("aaabccddsa"))  # Output: "a3b2c2dsa"
```

## LONGEST DUPLICATE SUBSTRING
Input - abaaabaababa  
Output - aaba

```python
def longestDupSubstring(s: str) -> str:
    # Generate the suffix array with (suffix, index) tuples
    suffixes = sorted((s[i:], i) for i in range(len(s)))
    
    def lcp(str1, str2):
        """Compute the longest common prefix of two strings."""
        min_len = min(len(str1), len(str2))
        for i in range(min_len):
            if str1[i] != str2[i]:
                return i
        return min_len
    
    longest_substring = ""
    
    # Compare consecutive suffixes to find the longest common prefix
    for i in range(1, len(suffixes)):
        common_len = lcp(suffixes[i-1][0], suffixes[i][0])
        if common_len > len(longest_substring):
            # Update the longest duplicated substring
            longest_substring = suffixes[i][0][:common_len]
    
    return longest_substring
```

```python
# Given a string s, return the length of the longest substring between two equal characters, excluding the two characters. 
# If there is no such substring return -1.
# A substring is a contiguous sequence of characters within a string.
def maxLengthBetweenEqualCharacters(s: str):
    max_length = -1
    longest_substring = ""
    n = len(s)
    
    # Iterate over each character
    for i in range(n):
        # Check subsequent characters for a match
        for j in range(i + 1, n):
            if s[i] == s[j]:
                # Calculate the length of the substring between the two equal characters
                length = j - i - 1
                # Update max_length and longest_substring if the current length is greater
                if length > max_length:
                    max_length = length
                    longest_substring = s[i + 1:j]  # Substring between i and j
                
    return max_length, longest_substring
```

## Sort chars in string based on frequency

```python
def frequencySort(s: str) -> str:
    # Step 1: Count the frequency of each character
    freq = {}
    for char in s:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    
    # Step 2: Sort the characters based on frequency in descending order
    sorted_chars = sorted(freq.items(), key=lambda x: -x[1])
    
    # Step 3: Rebuild the string by repeating each character according to its frequency
    result = ''.join([char * count for char, count in sorted_chars])
    
    return result
```

```python
# For a given string(str), remove all the consecutive duplicate characters.
# Example:
# Input String: "aaaa"
# Expected Output: "a"

# Input String: "aabbbcc"
# Expected Output: "abc"
def removeConsecutiveDuplicates(string):
    # Initialize an empty string to store the result
    result = ''
    
    # Start with the first character of the string
    i = 0
    while i < len(string):
        # Get the current character
        current_char = string[i]
        
        # Skip all consecutive duplicate characters
        j = i + 1
        while j < len(string) and string[j] == current_char:
            j += 1
        
        # Add the current character to the result string
        result += current_char
        
        # Move to the next non-duplicate character
        i = j
    
    return result

# Test the function with an example input
ans = removeConsecutiveDuplicates('aabbbcccdd')
print(ans)  # Output: "abcd"
```

```python
# For a given a string(str), find and return the highest occurring character.
# Example:
# Input String: "abcdeapapqarr"
# Expected Output: 'a'
# Since 'a' has appeared four times in the string which happens to be the highest frequency character, the answer would be 'a'.
# If there are two characters in the input string with the same frequency, return the character which comes first.
# Consider:
# Assume all the characters in the given string to be in lowercase always.

# Sample Input 1:
# abdefgbabfba
# Sample Output 1:
# b
def highest_occurring_char(s: str) -> str:
    # Initialize a dictionary to count the frequency of characters
    freq = {}
    
    # Count the frequency of each character in the string
    for char in s:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    
    # Initialize variables to track the character with the highest frequency
    highest_freq = -1
    highest_char = ''
    
    # Iterate over the string again to preserve the order of appearance
    for char in s:
        # If current character's frequency is greater than the highest frequency found so far
        if freq[char] > highest_freq:
            highest_freq = freq[char]
            highest_char = char
    
    return highest_char, highest_freq

# Sample Test Cases
print(highest_occurring_char("abdefgggggbabafba"))  # Output: 'b'
```

```python
# Given an integer N, the task is to check whether the given number is a pangram or not. 
# Note: A Pangram Number contains every digit [0- 9] at least once.
# Examples:
# Input : N = 10239876540022
# Output : Yes
# Explanation: N contains all the digits from 0 to 9. Therefore, it is a pangram.
# Input : N = 234567890
# Output : No
# Explanation: N doesn’t contain the digit 1. Therefore, it is not a pangram.

def numberPangram(N):
    # Initialize a list to keep track of digits seen (0 to 9)
    digits_seen = [False] * 10
    
    # Convert the number to string to iterate over each digit
    num_str = str(N)
    
    # Mark digits as seen
    for digit in num_str:
        digits_seen[int(digit)] = True
    
    # Check if all digits from 0 to 9 are seen
    return all(digits_seen)


# Driver Code
N = 239876540022
print("Yes" if numberPangram(N) else "No")

```

## RAIL FENCE ENCRYPTION

```python
def encrypt_rail_fence(message, num_rails=3):
    rails = ['' for _ in range(num_rails)]

    for i, char in enumerate(message):
        rail_index = i % num_rails
        rails[rail_index] += char

    encrypted_message = ''
    for rail in rails:
        encrypted_message += rail

    return encrypted_message

# Example usage:
user_input = input("Enter the string to encrypt: ")
encrypted_message = encrypt_rail_fence(user_input, 5)
print("Encrypted message:", encrypted_message)
```

## RAIL FENCE DECRYPTION

```python
def decrypt_rail_fence(encrypted_message, num_rails=3):
    rail_length = len(encrypted_message) // num_rails
    rails = ['' for _ in range(num_rails)]

    # Distribute characters to the rails
    for i in range(num_rails):
        if i < len(encrypted_message) % num_rails:
            rails[i] = encrypted_message[i * (rail_length + 1):(i + 1) * (rail_length + 1)]
        else:
            rails[i] = encrypted_message[i * rail_length + len(encrypted_message) % num_rails:
                                            (i + 1) * rail_length + len(encrypted_message) % num_rails]

    # Reconstruct the original message
    decrypted_message = ''
    for i in range(rail_length + 1):
        for j in range(num_rails):
            if i < len(rails[j]):
                decrypted_message += rails[j][i]

    return decrypted_message

# Example usage:
encrypted_input = input("Enter the string to decrypt: ")
decrypted_message = decrypt_rail_fence(encrypted_input, 5)
print("Decrypted message:", decrypted_message)
```

```python
def create_word_dict(decrypted_message):
    # Split the decrypted message into words
    words = decrypted_message.split()

    # Create a dictionary with the first character of each word as the key
    word_dict = {}
    for word in words:
        first_char = word[0].lower()  # Convert to lowercase for case-insensitivity
        if first_char in word_dict:
            word_dict[first_char].append(word)
        else:
            word_dict[first_char] = [word]

    # Sort the dictionary based on the number of elements in each list in descending order
    sorted_word_dict = dict(sorted(word_dict.items(), key=lambda x: len(x[1]), reverse=True))

    return sorted_word_dict

# Example usage:
decrypted_input = input("Enter Decrypted Message: ")
word_dict = create_word_dict(decrypted_input)

print("Output:", word_dict)
```

## MAGIC SQUARE
- The sum of elements vertically, horizontally and diagonally is same

```python
def is_magic_square(square):
    # Check if all rows have the same sum
    row_sums = []
    for row in square:
        row_sum = 0
        for element in row:
            row_sum += element
        row_sums.append(row_sum)
    if len(set(row_sums)) != 1:
        return False

    # Check if all columns have the same sum
    col_sums = []
    for j in range(3):
        col_sum = 0
        for i in range(3):
            col_sum += square[i][j]
        col_sums.append(col_sum)
    if len(set(col_sums)) != 1:
        return False

    # Check the sum of the main diagonal
    main_diag_sum = 0
    for i in range(3):
        main_diag_sum += square[i][i]
    if main_diag_sum != row_sums[0]:
        return False

    # Check the sum of the other diagonal
    other_diag_sum = 0
    for i in range(3):
        other_diag_sum += square[i][2 - i]
    if other_diag_sum != row_sums[0]:
        return False

    # If all conditions are met, it's a magic square
    return True

# Example usage:
# Replace the values in the square_list with your own 3x3 list
square_list_3x3 = [
    [2, 7, 6],
    [9, 5, 1],
    [4, 3, 8]
]

if is_magic_square(square_list_3x3):
    print("It's a magic square!")
else:
    print("It's not a magic square.")
```

## Here is an old puzzle question you can solve with a computer program. There is only one five-digit number n that is such that every one of the following ten numbers shares exactly one digit in common in the same position as n. Find n.
01265, 12171, 23257, 34548, 45970, 56236, 67324, 78084, 89872, 99414

```python
def common(a, b):
    count = []
    for i in range(5):
        if a[i] == b[i]:
            count.append(a[i])
    return len(count)

l = ['01265', '12171', '23257', '34548', '45970', '56236', '67324', '78084', '89872', '99414']

for i in range(10000, 100000):
    for x in l:
        if common(str(i), x) != 1:
            break
    else:
        print(i)
```

## STROBOGRAMMATIC NUMBERS

```python
def gen_strobogrammatic(n):
    """
    :type n: int
    :rtype: List[str]
    """
    result = helper(n, n)
    return result


def helper(n, length):
    if n == 0:
        return [""]
    if n == 1:
        return ["1", "0", "8"]
    middles = helper(n-2, length)
    result = []
    for middle in middles:
        if n != length:
            result.append("0" + middle + "0")
        result.append("8" + middle + "8")
        result.append("1" + middle + "1")
        result.append("9" + middle + "6")
        result.append("6" + middle + "9")
    return result

print("n = 2: \n",gen_strobogrammatic(2))
print("n = 3: \n",gen_strobogrammatic(3))
print("n = 4: \n",gen_strobogrammatic(4))
```

## longest prefix 

```python
def longest_common_prefix(str_list):
    if not str_list:
        return -1

    str_list.sort()

    prefix = ""
    for i in range(len(str_list[0])):
        char = str_list[0][i]
        prefix_match = True

        for s in str_list:
            if s[i] != char:
                prefix_match = False
                break

        if prefix_match:
            prefix += char
        else:
            break

    return prefix if prefix else -1

# Example usage
input_list1 = ["lessonplan", "lesson", "lees", "length"]
result1 = longest_common_prefix(input_list1)
print("Input list:", input_list1)
print("The longest Common Prefix is:", result1)

input_list2 = ["python", "pythonprogramming", "pythonlist"]
result2 = longest_common_prefix(input_list2)
print("\nInput list:", input_list2)
print("The longest Common Prefix is:", result2)

input_list3 = ["lessonplan", "lesson", "ees", "length"]
result3 = longest_common_prefix(input_list3)
print("\nInput list:", input_list3)
print("The longest Common Prefix is:", result3)
```
