## Python Modules: Overview
There are actually three different ways to define a module in Python:

- A module can be written in Python itself.
- A module can be written in C and loaded dynamically at run-time, like the re (regular expression) module.
- A built-in module is intrinsically contained in the interpreter, like the itertools module.

# PACKAGES

## COMPARISON


| **Feature**      | **Module**                                     | **Library**                                       | **Package**                                       | **Framework**                                     |
|------------------|------------------------------------------------|---------------------------------------------------|---------------------------------------------------|---------------------------------------------------|
| **Definition**   | A single Python file containing code.           | A collection of modules that form a set of tools. | A collection of modules organized in a directory hierarchy. | A pre-built, organized structure for building applications. |
| **Organization** | Standalone unit.                                | Group of related modules.                         | Group of related modules and sub-packages.         | Pre-structured architecture for application development. |
| **Importing**    | Imported using the `import` statement.          | Imported using the `import` statement.            | Imported using the `import` statement.            | Integrated structure, often involves specific patterns or methodologies. |
| **Example**      | `import math`                                  | `import os`                                      | `from my_package import module1, module2`        | Django, Flask (web frameworks), TensorFlow (machine learning framework) |
| **Use Case**     | Code organization and reuse.                    | Providing reusable functionalities.              | Organizing and managing related modules.         | Building applications based on a specific architecture. |
| **Directory**    | Single file.                                   | Group of related files in a directory.           | Directory with `__init__.py` and related module files. | Pre-defined directory structure and components.   |

## SUBPACKAGE

you can use a relative import, where .. refers to the package one level up. From within mod3.py, which is in subpackage sub_pkg2,

- .. evaluates to the parent package (pkg), and
- ..sub_pkg1 evaluates to subpackage sub_pkg1 of the parent package.

```
from .. import sub_pkg1
print(sub_pkg1)

from ..sub_pkg1.mod1 import foo
foo()
```

# `os` module

**OS Module:**

The `os` module in Python provides functions for interacting with the operating system. Here are some key functions:

## 1. `getcwd()`: Returns the current working directory.

   ```python
   import os
   current_directory = os.getcwd()
   print("Current Directory:", current_directory)
   ```

## 2. `chdir(path)`: Changes the current working directory to the specified path.

   ```python
   import os
   os.chdir('/path/to/new/directory')
   ```

## 3. `mkdir(path)`: Creates a new directory at the specified path.

   ```python
   import os
   os.mkdir('/path/to/new/directory')
   ```

## 4. `listdir(path)`: Returns a list of files and directories in the specified path.

   ```python
   import os
   contents = os.listdir('/path/to/directory')
   print("Contents:", contents)
   ```

## 5. `remove(file_path)`: Removes (deletes) the file at the specified path.

   ```python
   import os
   os.remove('/path/to/file.txt')
   ```

## 6. `rmdir(directory_path)`: Removes (deletes) the empty directory at the specified path.

   ```python
   import os
   os.rmdir('/path/to/empty_directory')
   ```
