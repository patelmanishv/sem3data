```A jar of Halloween candy contains an unknown amount of candy and if you can guess exactly
how much candy is in the bowl, then you win all the candy. You ask the person in charge the
following: If the candy is divided evenly among 5 people, how many pieces would be left
over? The answer is 2 pieces. You then ask about dividing the candy evenly among 6 people,
and the amount left over is 3 pieces. Finally, you ask about dividing the candy evenly among
7 people, and the amount left over is 2 pieces. By looking at the bowl, you can tell that there
are less than 200 pieces. Write a program to determine how many pieces are in the bowl.
```

````python
def find_candy_count():
    # Loop through all possible numbers less than 200
    for x in range(200):
        # Check if the current number satisfies all conditions
        if x % 5 == 2 and x % 6 == 3 and x % 7 == 2:
            return x  # Return the correct number of candies

# Find and print the number of candies
candy_count = find_candy_count()
print(f"The number of pieces in the bowl is: {candy_count}")
````

```
The number 1961 reads the same upside-down as right-side up. Print out all the numbers between 1 and 100000 that read the same upside-down as right-side up.
```

````python
def upside_down_digit(digit):
    # Return the upside-down equivalent of the digit
    if digit == 0:
        return 0
    elif digit == 1:
        return 1
    elif digit == 6:
        return 9
    elif digit == 8:
        return 8
    elif digit == 9:
        return 6
    else:
        return -1  # Invalid digit for upside-down

def is_upside_down_same(num):
    original = num
    flipped = 0
    power_of_10 = 1  # To keep track of the position of flipped digits

    while num > 0:
        digit = num % 10
        flipped_digit = upside_down_digit(digit)

        # If the digit is not valid for upside-down, return False
        if flipped_digit == -1:
            return False

        # Build the flipped number
        flipped = flipped * 10 + flipped_digit
        num //= 10

    # Check if the flipped number is the same as the original
    return flipped == original

# Iterate over all numbers between 1 and 100000
for num in range(1, 100001):
    if is_upside_down_same(num):
        print(num)
````

```
The number 99 has the property that if we multiply its digits together and then add the sum of its digits to that, we get back to 99. That is, (9 · 9) + (9 + 9) = 99. Write a program to find all of the numbers less than 10000 with this property. (There are only nine of them.)
```

````python
def calculate_digit_sum_and_product(num):
    digit_sum = 0
    digit_product = 1
    temp = num

    # If the number is 0, handle it separately
    if temp == 0:
        return 0, 0

    while temp > 0:
        digit = temp % 10
        digit_sum += digit
        digit_product *= digit
        temp //= 10

    return digit_sum, digit_product

def find_special_numbers(limit):
    for num in range(10, limit):  # Numbers below 10 are trivial cases
        digit_sum, digit_product = calculate_digit_sum_and_product(num)
        if digit_sum + digit_product == num:
            print(num)

# Find and print all numbers less than 10000 with the special property
find_special_numbers(10000)
````

````python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def next_prime(current):
    num = current + 1
    while True:
        if is_prime(num):
            return num
        num += 1

def print_upper_half_prime_diamond(n):
    prime = 2  # Start with the first prime number
    
    # Top half of the diamond
    for i in range(n):
        # Print leading spaces
     
        
        # Print prime numbers for the current row
        for _ in range(i + 1):
            print(prime, end=" ")
            prime = next_prime(prime)  # Get the next prime number
        
        print()  # Move to the next line

# Adjust 'n' to control the size of the diamond's upper half
n = 6
print_upper_half_prime_diamond(n)
````

## PYTHAGORIAN TRIPLET PROBLEM

````python
# Function to manually calculate integer square root
def integer_sqrt(n):
    guess = 1
    while guess * guess <= n:
        guess += 1
    return guess - 1

# Function to find and print Pythagorean triplets
def find_pythagorean_triplets(limit):
    # Iterate over possible values of a and b
    for a in range(1, limit):
        for b in range(a, limit):  # Start from 'a' to avoid duplicate pairs
            c_square = a * a + b * b
            
            # Manually compute the integer square root of c_square
            c = integer_sqrt(c_square)
            
            # Check if c^2 equals a^2 + b^2 and c is <= limit
            if c * c == c_square and c <= limit:
                print(f"({a}, {b}, {c})")

# Example usage: Find Pythagorean triplets up to 100000
limit = 20
find_pythagorean_triplets(limit)
````

## Find smallest integer that satisfy following conditions
- When the rightmost digit is moved to the very left, the new number so formed must be exactly 1.5 times the original number.

````python
def smallest_integer():
    n = 1  # Start checking from the smallest positive integer
    while True:
        # Extract the rightmost digit
        rightmost_digit = n % 10
        
        # Remove the rightmost digit from the original number
        rest_of_number = n // 10
        
        # Calculate the number of digits in the original number
        k = 0
        temp = n
        while temp > 0:
            temp //= 10
            k += 1
            
        # Create the new number by moving the rightmost digit to the front
        new_number = rightmost_digit * (10 ** (k - 1)) + rest_of_number
        
        # Check if the new number is 1.5 times the original number
        if new_number == 1.5 * n:
            return n
        
        n += 1  # Move to the next integer

# Find and print the smallest positive integer
result = smallest_integer()
print(result)
````

## sheep rabbit problem

````python
def solve_sheep_rabbit_problem(total_heads, total_legs):
    # Check for the possibility of the solution first
    if total_legs % 2 != 0 or total_heads > total_legs // 2:
        return "No solution"

    # Use algebra to solve the equations
    # From x + y = total_heads and 4x + 2y = total_legs
    # y = total_heads - x
    for sheep in range(total_heads + 1):
        rabbits = total_heads - sheep
        if 4 * sheep + 2 * rabbits == total_legs:
            return f"Sheep: {sheep}, Rabbits: {rabbits}"

    return "No solution"

# Example usage:
total_heads = 10
total_legs = 26
result = solve_sheep_rabbit_problem(total_heads, total_legs)
print(result)
````

## find third maximum and average

````python
def find_third_maximum_and_average():
    # Initialize three variables to store the maximum values
    first_max = second_max = third_max = float('-inf')
    total_sum = 0  # Variable to store the sum of all numbers

    for i in range(5):  # Loop to input 5 numbers
        num = int(input(f"Enter number {i+1}: "))
        total_sum += num  # Add to the total sum
        
        # Update the three maximum values accordingly
        if num > first_max:
            third_max = second_max
            second_max = first_max
            first_max = num
        elif num > second_max:
            third_max = second_max
            second_max = num
        elif num > third_max:
            third_max = num

    # Calculate the sum of the remaining four numbers
    sum_of_rest = total_sum - third_max
    
    # Calculate the average of the remaining four numbers
    average = sum_of_rest / 4
    
    return average, third_max

# Example usage:
average, third_max = find_third_maximum_and_average()
print(f"The third maximum number is: {third_max}")
print(f"The average of the remaining four numbers is: {average}")
````

## Is perfect square?

````python
def is_perfect_square(num):
    if num < 1:
        return False
    
    odd_number = 1
    while num > 0:
        num -= odd_number
        odd_number += 2  # Move to the next odd number
    
    return num == 0

# Example usage:
number = int(input("Enter a number: "))
if is_perfect_square(number):
    print(f"{number} is a perfect square.")
else:
    print(f"{number} is not a perfect square.")
````

## find missing number from incremental user input

````python
def find_missing_number():
    print("Enter 5 numbers in increasing order:")
    
    # Read the first number
    first_num = int(input("Number 1: "))
    actual_sum = first_num  # Initialize the actual sum with the first number
    last_num = first_num     # Set last_num to the first input number

    # Read the next 4 numbers
    for i in range(2, 6):
        num = int(input(f"Number {i}: "))
        actual_sum += num  # Accumulate the sum
        last_num = num      # Update the last number to the current input

    # Calculate the expected sum from first_num to last_num
    expected_sum = 0
    for i in range(first_num, last_num + 1):  # Loop from first_num to last_num
        expected_sum += i  # Accumulate the expected sum
    
    # The missing number is the difference
    missing_number = expected_sum - actual_sum
    
    return missing_number

# Example usage:
missing_number = find_missing_number()
print(f"The missing number is: {missing_number}")
````

```
Given an integer num, return the number of steps to reduce it to zero.

In one step, if the current number is even, you have to divide it by 2, otherwise, you have to subtract 1 from it.
```

## to find number of steps to reduce n to zero - conditions are:
- If number is even, divide it by 2, if number is odd subtract 1

````python
def numberOfSteps(num):
    steps = 0  # Initialize a counter for the number of steps

    while num > 0:
        if num % 2 == 0:  # Check if the number is even
            num //= 2  # Divide the number by 2
        else:
            num -= 1  # Subtract 1 from the number
        steps += 1  # Increment the step count

    return steps  # Return the total number of steps

# Driver code
if __name__ == '__main__':
    # Example input
    n = 8
    result = numberOfSteps(n)
    
    # Print the result
    print(result)  # Output will be the number of steps to reduce n to zero
````
